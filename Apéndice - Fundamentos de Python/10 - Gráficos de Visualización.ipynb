<h1 style="text-align: justify;">Gráficos de visualización</h1>
<p style="text-align: justify;">

Los gráficos de visualización son representaciones visuales de datos que ayudan a comprender patrones, tendencias y relaciones. En Python, se pueden crear utilizando bibliotecas como Matplotlib, Seaborn, Plotly, entre otras, que ofrecen una amplia gama de opciones para crear gráficos interactivos y estáticos.
    
</p>
</div>

### Matplotlib

```python
import matplotlib.pyplot as plt

# Gráfico de lineas
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
plt.plot(x, y)
plt.grid()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfico de líneas")
plt.show()
```

```python
import numpy as np

x = np.random.random(size=(100,))
y = np.random.random(size=(100,))
plt.scatter(x, y)
plt.grid()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfico de dispersión")
plt.show()
```

### Seaborn

```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Cargar los datos
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)
iris_df
```

```python
# Crear gráfico
sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="species", data=iris_df)
plt.title("Gráfico de Dispersión - Iris Dataset")
plt.show()
```

### Pandas

```python
datos = np.random.normal(size=(100_000,))
df = pd.DataFrame(datos)
df.plot(kind="hist", bins=100)
plt.grid()
plt.title("Histograma con Pandas")
plt.show()
```

### Alternativas:

Existe una gran cantidad de librerías con las que puedes realizar gráficos. Las más famosas son matplotlib y seaborn.

Librerías opcionales:


- Plotly: Permite la creación de gráficos interactivos y dashboards. Es especialmente útil para visualizaciones web y es compatible con una amplia variedad de tipos de gráficos.

- Bokeh: Ofrece una forma elegante de construir gráficos interactivos y aplicaciones web visuales. Es especialmente útil cuando se necesita interactuar con los gráficos en un navegador web.

- Altair: Una biblioteca declarativa que permite crear gráficos interactivos utilizando una sintaxis simple y concisa. Está integrada con Pandas y es fácil de aprender.

- Dash (por Plotly): Es una biblioteca para crear aplicaciones web interactivas basadas en Python. Permite la creación de dashboards interactivos.

- ggplot (por yhat): Basada en el popular sistema de gráficos ggplot2 de R, proporciona una interfaz similar para crear gráficos gramaticales en Python.

- NetworkX: Especializada en la visualización y análisis de gráficos y redes complejas. Es útil para representar relaciones y estructuras de redes.

- Holoviews: Una biblioteca que facilita la exploración y visualización de datos, proporcionando una abstracción simple para gráficos complejos.
